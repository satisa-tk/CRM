# **Load Data**

```python
df = pd.read_excel('Customer Behaviors (Responses).xlsx')
df = df[df.columns[1:]]
df.head()
```

# **Rename Columns** 

```python
#Replace Columns name
df.columns = df.columns.str.replace(r'(คุณมีความสนใจในสิ่งเหล่านี้มากน้อยเพียงใด)',
                              'interest').str.replace(r'(คุณบริโภคสิ่งเหล่านี้บ่อยขนาดไหน)',
                              'consume').str.replace(r'(คุณเกิดวันที่เท่าไหร่)','dob').str.replace(r'(เพศ)','gender',regex=True).str.strip()
                              
#Rename gender
df['gender'] = df['gender'].map({'หญิง':'Female', 'ชาย':'Male', 'ไม่ต้องการระบุ': 'X'})
df.head()
```

# **likert scale**

```python
interest = df.replace({'ไม่สนใจอย่างมากที่สุด': 1, 'ไม่สนใจอย่างมาก': 2, 'ไม่สนใจ': 3, 'เฉยๆ': 4, 'สนใจ': 5, 'สนใจอย่างมาก': 6, 'สนใจอย่างมากที่สุด': 7}, inplace=True)
consume = df.replace({'แทบไม่ได้บริโภคเลย': 1, 'หลายเดือนครั้ง': 2, 'เดือนละครั้ง': 3, 'เดือนละ 2-3 ครั้ง': 4, 'อาทิตย์ละครั้ง': 5, 'แทบทุกวัน': 6}, inplace=True)
df.head()
```
